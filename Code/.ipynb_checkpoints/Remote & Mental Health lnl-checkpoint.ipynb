# Remote Work and Mental Health


How do the number of hours worked per week and regional differences impact employee productivity in remote work environments in North America vs. Asia?

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv('../resources/Impact_of_Remote_Work_on_Mental_Health.csv')
df.head()
```

```python
North_AmericaDF = df[df["Region"] == 'North America']  
#print(North_AmericaDF)
```

```python
AsiaDF = df[df["Region"] == 'Asia']  
#print(AsiaDF)
```

```python
# Filter for North America
north_america_data = df[df['Region'] == 'North America'].copy()
# Bin hours worked into ranges
bins = [0, 20, 40, 60, 80]
bin_labels = ['0-20', '21-40', '41-60', '61-80']
north_america_data['Hours_Worked_Bins'] = pd.cut(
    north_america_data['Hours_Worked_Per_Week'], bins=bins, labels=bin_labels, right=False
)
# Group data by hours worked bins and productivity change
grouped_data = north_america_data.groupby(['Hours_Worked_Bins', 'Productivity_Change'], observed=True).size().unstack(fill_value=0)
# Plot a binned bar chart
grouped_data.plot(kind='bar', figsize=(12, 6), width=0.8, color = ['#8c96c6', '#74c476', '#6baed6'])
# Customize the chart
plt.title("Productivity Change vs. Hours Worked (North America)", fontsize=14)
plt.xlabel("Hours Worked Per Week", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.legend(title="Productivity Change", labels=['Decrease', 'No Change', 'Increase'])
# Show the plot
plt.tight_layout()
plt.show()

```

## North America Analysis:
On average in North America employees who report working between 21-40 hours per week report the highest productivity increases. The employees who reported working 61-80 hours per week reported mostly no change in productivity and very little in increased productivity. This can depict problems like burn out and reduced efficiancy.



```python
# Filter for Asia
asia_data = df[df['Region'] == 'Asia'].copy()
# Bin hours worked into ranges
bins = [0, 20, 40, 60, 80]
bin_labels = ['0-20', '21-40', '41-60', '61-80']
asia_data['Hours_Worked_Bins'] = pd.cut(
    asia_data['Hours_Worked_Per_Week'], bins=bins, labels=bin_labels, right=False
)
# Group data by hours worked bins and productivity change
grouped_data = asia_data.groupby(['Hours_Worked_Bins', 'Productivity_Change'], observed=True).size().unstack(fill_value=0)
# Plot a binned bar chart
grouped_data.plot(kind='bar', figsize=(12, 6), width=0.8, color = ['#8c96c6', '#74c476', '#6baed6'])
# Customize the chart
plt.title("Productivity Change vs. Hours Worked (Asia)", fontsize=14)
plt.xlabel("Hours Worked Per Week", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.legend(title="Productivity Change", labels=['Decrease', 'No Change', 'Increase'])
# Show the plot
plt.tight_layout()
plt.show()
```

## Asia Analysis :

The employees who reported working between 41-60 hours per week reported a larger decrease in productivity. In Asia the employees who reported working 61-80 reported similar levels of productivity.


## Regional Comparison :

The bar charts analyzing Productivity Change versus Hours Worked Per Week for North America and Asia reveal distinct regional trends. In North America, productivity changes are relatively balanced across the 21-40 and 41-60 hour bins, with many employees reporting "No Change" or an "Increase" in productivity. In contrast, Asia shows a greater occurrence of "Decrease" in productivity within these same hour ranges, indicating potential challenges associated with longer working hours in remote settings. 

Both regions have minimal representation in the 61-80 hour bin, suggesting that few employees work extensive hours. Despite this, productivity outcomes in this category vary significantly. North America shows more stable and positive productivity trends, while Asia has more variation and challenges. This could be due to differences in culture, workplace practices, or resources. These results suggest the importance of creating remote work policies and support systems that are tailored to the unique needs of each region.

```python
# Replace the values
north_america_data['Productivity_Change'] = north_america_data['Productivity_Change'].replace({
    'Increase': 1,
    'No Change': 0,
    'Decrease': -1
})

# Convert the column to float explicitly
north_america_data['Productivity_Change'] = north_america_data['Productivity_Change'].astype(float)
```

```python
plt.figure(figsize=(12, 6))
north_america_data.boxplot(
    column='Productivity_Change',
    by='Hours_Worked_Bins',
    grid=False,
    patch_artist=True,
    boxprops=dict(facecolor='lightblue', color='black'),
    medianprops=dict(color='red'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    flierprops=dict(marker='o', color='orange', alpha=0.5)
)
plt.title("Productivity Change by Hours Worked in North America", fontsize=14)
plt.suptitle("")
plt.xlabel("Hours Worked Per Week", fontsize=12)
plt.ylabel("Productivity Change", fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## North America Analysis :

The box plot analyzes the relationship between hours worked per week and productivity changes in North America, with bins representing 21-40, 41-60, and 61-80 hours. Across all bins, the median productivity change is consistently 0, indicating that most employees report no change in productivity regardless of hours worked. The interquartile range (IQR) spans from -1 (Decrease) to 1 (Increase), showing that while some employees experience increased productivity, others report a decrease, reflecting variability in individual experiences. The distributions across the bins are highly similar, suggesting that hours worked per week may not significantly influence productivity change.
The absence of data for the 0-20 hours bin further reinforces that this range is not representative in the dataset. Other factors, such as stress levels or work-life balance, may have a greater impact on productivity than hours worked alone.

```python
# Replace values and explicitly convert to numeric type
asia_data['Productivity_Change'] = asia_data['Productivity_Change'].replace({
    'Increase': 1,
    'No Change': 0,
    'Decrease': -1
}).astype(float)  
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Ensure proper updates to 'Productivity_Change' and 'Hours_Worked_Bins'
asia_data.loc[:, 'Productivity_Change'] = asia_data['Productivity_Change'].replace(
    {'Decrease': -1, 'No Change': 0, 'Increase': 1}
)

asia_data.loc[:, 'Hours_Worked_Bins'] = pd.cut(
    asia_data['Hours_Worked_Per_Week'],
    bins=[0, 20, 40, 60, 80],
    labels=['0-20', '21-40', '41-60', '61-80'],
    right=False,
    include_lowest=True
)

# Verify if there is valid data to plot
if not asia_data.empty and not asia_data['Hours_Worked_Bins'].isnull().all():
    # Create the box plot
    plt.figure(figsize=(12, 6))
    asia_data.boxplot(
        column='Productivity_Change',
        by='Hours_Worked_Bins',
        grid=False,
        patch_artist=True,
        boxprops=dict(facecolor='lightblue', color='black'),
        medianprops=dict(color='red'),
        whiskerprops=dict(color='black'),
        capprops=dict(color='black'),
        flierprops=dict(marker='o', color='orange', alpha=0.5)
    )

    # Add customizations
    plt.title("Productivity Change by Hours Worked in Asia", fontsize=14)
    plt.suptitle("")  # Remove the default subplot title
    plt.xlabel("Hours Worked Per Week", fontsize=12)
    plt.ylabel("Productivity Change", fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()

    # Show the plot
    plt.show()
else:
    print("No valid data available to plot for Asia.")
```

## Asia Analysis :

The results indicate that most employees report no change in productivity, with a consistent median of 0 across all bins. However, the 61-80 hours bin shows greater variability, including notable decreases in productivity, while the 0-20 hours bin lacks sufficient data for analysis. These findings suggest that hours worked per week may have minimal impact on productivity, emphasizing the need to investigate other factors, such as mental health resources, work-life balance, and organizational support, to better understand and improve productivity in remote work settings.

## Regional Comparison :

The analysis reveals that in both regions, the median productivity change is 0 across all hour ranges (21-40, 41-60, and 61-80), indicating that most employees experience no significant change in productivity regardless of hours worked. However, Asia shows greater variability, especially in the 61-80 hour range, where some employees report significant drops in productivity. In contrast, North America displays more consistent patterns with no major outliers. These findings suggest that while hours worked provide some insights, other factors—such as stress levels and work-life balance—likely play a more significant role in influencing productivity outcomes.

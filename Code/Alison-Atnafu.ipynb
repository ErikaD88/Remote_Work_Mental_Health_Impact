```python
# Importing necessary libraries
import pandas as pd            
import numpy as np 
import matplotlib.pyplot as plt 
from pathlib import Path 
```

```python
# locating the data set 
mental_health_data = Path("resource/Impact_of_Remote_Work_on_Mental_Health.csv")
```

```python
# Reads the CSV file located at the path stored in the mental_health_data variable.
complete_health_data = pd.read_csv(mental_health_data)

complete_health_data.head()
```

```python
# Accesses the "Region" column in the complete_health_data DataFrame.
north_america_data = complete_health_data[complete_health_data["Region"].str.strip() == "North America"]
north_america_data.head()
```

```python
# This pull the number of employees in North America 
emp_north_america = north_america_data["Region"].count() # Count north America Employees 
print (f"Number of employees in North America: {emp_north_america}")
```

```python
# Accessing the remot workers in North America 
NA_remot_employee = north_america_data [north_america_data["Work_Location"] == "Remote"]

NA_remot_employee.head()
```

```python
# This will count the number of remot workers in North America
remote_worker_count = NA_remot_employee["Work_Location"].value_counts().get("Remote")
print(f"Number of remote workers in North America: {remote_worker_count}")
```

```python
# This display the number of mental health condition based on types 
health_condition = NA_remot_employee["Mental_Health_Condition"].value_counts()
health_condition
```

```python
# Calculate the total number of employees with a mental health condition
total_condition = health_condition.sum()
print(f"The number of employees with mental health conditions: {total_condition}")
```

```python
# Generate a sequence of colors from the 'viridis' colormap for the number of categories in health_condition
colors = plt.cm.viridis(np.linspace(0, 1, len(health_condition)))

# Plot the pie chart with the generated colors
health_condition.plot(
    kind="pie",  # Create a pie chart
    autopct="%1.1f%%",  # Display percentage values with one decimal place
    colors=colors,  # Set the random colors of the pie slices
)

# Add a title to the chart
plt.title("Distribution of Mental Health Condition Among Remote Workers in North America")

plt.ylabel("") # This remove the y-axis label.
plt.tight_layout()
# Save the pie chart as an image (
plt.savefig("mental_health_condition_distribution.png", format="png")
# Display the pie chart
plt.show()
```

```python
# Generate colors from the 'viridis' colormap for the number of unique conditions
colors = plt.cm.viridis(np.linspace(0, 1, len(health_condition)))

# Create the bar chart with the generated colors
health_condition.plot(
    kind="bar",  # Create a bar chart
    color=colors,  # Set colors of the bars
)

# Set the x-axis label
plt.xlabel("Mental Health Condition")  # Label for x-axis: Mental Health Conditions (categories)

# Set the y-axis label: this explains the number of employees with each condition
plt.ylabel("# of Employees with Mental Health Condition")

plt.xticks(rotation=0) # Adding rootation for the x label ticks
# Set the title for the chart with corrected spelling of "America"
plt.title("Prevalence of Mental Health Conditions Among Remote Workers in North America")

# Adjust the layout to prevent clipping of labels
plt.tight_layout()

# Save the plot 
plt.savefig("prevalence_of_mental_health_condition.png", format='png')
plt.show()
```

```python

employee_gender = NA_remot_employee.groupby("Mental_Health_Condition")["Gender"].value_counts().reset_index()
employee_gender
```

```python
# Aggregate the data to get the total count of each gender
gender_totals = employee_gender.groupby(["Mental_Health_Condition", "Gender"])["count"].sum()

# Plot the pie chart
# Plot the data as a pie chart
gender_totals.plot(
    kind="pie",  
    autopct="%1.1f%%",  # Displays the percentage of each segment with one decimal place
    startangle=90,  # Rotates the start angle of the pie chart by 90 degrees 
    figsize=(8, 8),  # Sets the size of the plot to be 8 inches by 8 inches
    
    # Generates a random color from the 'viridis' colormap based on unique categories
    colors=plt.cm.viridis(np.linspace(0, 1, len(gender_totals)))  
)

# Add a title
plt.title("Proportion of Mental Health Conditions by Gender")
plt.ylabel("")  # Remove default y-axis label
plt.tight_layout()
plt.savefig("Mental Health Condition Proportions by Gender.png", format="png")
plt.show()
```

```python
# To reaorganize the data and show the prevalence rate of mental health condition among different gender categories 
# we used pivot function to extract the desired data from employee_gender dataframe 
pivot_table = employee_gender.pivot(index="Mental_Health_Condition", columns="Gender", values="count")

# Creating a bar chart from the pivot_table DataFrame.
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)

# Add labels and title
plt.title("Mental Health Conditions Across Genders")
plt.xlabel("Mental Health Condition")
plt.ylabel("Count of Employees")
plt.xticks(rotation=0)
plt.legend(loc='upper left', fontsize='small')
plt.tight_layout()
plt.savefig("Mental_Health_Across_Genders.png", format='png')
# Display the chart
plt.show()
```
